```python
# Llevar datos al database. Mongo trabaja con colecciones, que son como las tablas en MySQL. Si tengo un dabase tambien debo tener mis colecciones

# No es necesario, pero es recomendado tener una copia del df original
df = df_customers.copy()

# Mongo almacena documetos por ello hay que almacenar el df como un diccionario para almacenarlo en un JSON que es compatible con Mongo
df_to_dict = df.to_dict("records") # records = parametros para indica registros o filas sea los elementos del diccionario

# Mandar datos a Mongo
'''
insert_many = Para insertar un dataframe completo
insert_one = Para insertar un registro del dataframe

dbname["collection_name"].insert_many(records_dict)
'''

dbname["customers"].insert_many(df_to_dict)
```

```python
# Ya enviamos datos hacia MongoDB, ahora vamos a traer datos desde MongoDB. Se puede usar la misma conexión pero usamos otros metodos

# Almacenar la collección de Mongo en una variable
collection_customers = dbname["customers"]

# Traer los datos desde Mongo
customers = collection_customers.find({})
```
