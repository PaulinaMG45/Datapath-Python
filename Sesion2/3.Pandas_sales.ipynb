```python
import pandas as pd
```

```python
'''
UnicodeDecodeError: 'utf-8' codec can't decode byte 0x84 in position 5327: invalid start byte

Este error ocurre cuando el archivo no está guardado en utf-8 y es necesario ajustar el encoding

Para saber el enconding (chatset)

¿Qué es Charset?
"Charset" es una abreviación de "character set" (conjunto de caracteres). Un charset es una colección de caracteres y un mapeo entre esos caracteres y sus valores numéricos correspondientes. Los charsets son fundamentales en la computación porque permiten la representación y manipulación de texto en diferentes idiomas y sistemas.

Función y Uso de Charset
Representación de Texto:

Los charsets definen cómo se representan los caracteres de texto en términos de bytes. Esto es esencial para almacenar y transmitir texto correctamente.
Codificación de Caracteres:

Un charset incluye un esquema de codificación que especifica cómo convertir cada carácter en una secuencia de bytes y viceversa. Ejemplos de esquemas de codificación son ASCII, UTF-8, UTF-16, ISO-8859-1, etc.


Para conocer el chatset de un archivo:
- Se debe estar en la carpeta donde está almacenado
- Mirar tipo de archivo file file.extension
- Mirar charset file -i file.ext

'''

df_sales = pd.read_csv("./data/sales_data_sample.csv",encoding='ISO-8859-1')
df_sales
```

```python
# Agrupaciones

# agrupacion con una sola variable
df_sales[['COUNTRY','PRICEEACH']].groupby('COUNTRY').mean() 

# agrupacion con varias variables
df = df_sales[['COUNTRY','PRODUCTCODE','PRICEEACH']].groupby(['COUNTRY','PRODUCTCODE']).mean() 
df
```

```python
data = {
    'A':[1,4,3,2]
    ,'B':[0,4,5,7]
    ,'C':[3,5,1,8]
}

df1 = pd.DataFrame(data)
df1
```

```python
data = {
    'C':[1,2,3,4,5]
    ,'D':[3,4,6,8,2]
    ,'E':[3,4,5,2,1]
}

df2 = pd.DataFrame(data)
df2
```

```python
# Concatenar a lo largo de fila = axis = 0
# Es util cuando tenemos muchos archivos con la misma estructura y necesitamos concatenarlos, leemos los df y obtenemos un df final con todos 
pd.concat((df1,df2), axis=0)
```

```python
# Concatenar a lo largo de columna = axis = 1

pd.concat((df1,df2), axis=1)
```

```python
# inner join

pd.merge(df1,df2,on='C')
```

```python
# inner join - desde el mismo dataframe

df1.merge(df2,on='C')
```

```python
# left join - desde el mismo dataframe

df1.merge(df2,on='C',how='left')
```

```python
# right join - desde el mismo dataframe

df1.merge(df2,on='C',how='right')
```

```python
# join - desde el mismo dataframe cuando los nombres de las llaves para unir son diferentes

df1.merge(df2,left_on='C',right_on='D',how='inner')
```
