```python
# Librerias para generar graficos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Grafico de linea - mostrar relacion entre dos variables continuas, tendencias en el tiempo y mostrar datos
# continuos series temporales

x = [1,2,3,4,5]
y = [2,3,5,7,9]


plt.plot(x,y)

# Personalizacion grafico
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de líneas')

# Para mostrar solo el gráfico sin la descripción del gráfico creado
plt.show()
```

```python
# Grafico de dispersion - Util para mostrar relacion de variables continuas, correlaciones o patrones dentre variables
# es usual para visualizar datos en analsis exploratorio y ciencia de datos para validar correlaciones


# Grafico de linea - mostrar relacion entre dos variables continuas, tendencias en el tiempo y mostrar datos
# continuos series temporales

x = [1,2,3,4,5]
y = [2,3,5,7,9]


plt.scatter(x,y)

# Personalizacion grafico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico de dispersión')

# Para mostrar solo el gráfico sin la descripción del gráfico creado
plt.show()
```

```python
# Histogramas - Permiten ver la distribuciónd de una variable numerica, ver la distribución de los datos en diferentes
# intervalos. También permiten ver valores atípicos

# Para generar datos
import numpy as np

datos = np.random.rand(1000) # retorna una lista de 1000 numeros

plt.hist(datos,bins=10,color='skyblue',edgecolor='black') # bin cantidad de intervalos, color y edgecolor para personalizacion

# Personalizacion grafico
plt.xlabel('Eje X')
plt.ylabel('Eje Y') # Frecuencia
plt.title('Histograma')

# Para mostrar solo el gráfico sin la descripción del gráfico creado
plt.show()
```

```python
# Grafico de barras - Util para comparar categorias o grupos

categoria = ['A','B','C']
valor = [150,230,180]

plt.bar(categoria,valor,color='green',edgecolor='black')

# Personalizacion grafico
plt.xlabel('categorias')
plt.ylabel('valor') 
plt.title('Grafico de barras')

# Para mostrar solo el gráfico sin la descripción del gráfico creado
plt.show()
```
